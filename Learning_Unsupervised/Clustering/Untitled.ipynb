```python
df=df.to_numpy()
```

```python
def kmeans(df,k,iterations=300):
    clusters=df[np.random.choice(len(df),size=(k))]
    j=0
    while (j<=iterations):
        old_clusters=clusters.copy()
        assignments=[]
        for i in df:
            distances = np.linalg.norm(clusters - i, axis=1)
            cluster_index = np.argmin(distances)
            assignments.append(cluster_index)
            
        assignments=np.array(assignments)
        for i in range(k):
            cluster_points = df[assignments == i]
            if len(cluster_points) > 0:
                clusters[i] = cluster_points.mean(axis=0)
            else:
                clusters[i] = df[np.random.choice(len(df))]

            clusters[i]=np.mean(df[assignments==i])
            
        shift = np.linalg.norm(clusters - old_clusters)
        if shift < 1e-40:
            print("Converged after",j,"iterations")
            return clusters, assignments   
        j+=1
```

```python
clusters, assignments = kmeans(df,3)
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df)
reduced_clusters = pca.transform(clusters)

plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=assignments, cmap='viridis')
plt.scatter(reduced_clusters[:, 0], reduced_clusters[:, 1], c='red', s=200, marker='X')
plt.title('K-Means Clustering (PCA-Reduced)')
plt.show()
```

```python
def clustering_accuracy(assignments, sol, k):
    best_accuracy = 0
    best_mapping = None

    for perm in permutations(range(k)):
        mapped = np.array([perm[label] for label in assignments])
        acc = np.mean(mapped == sol)
        if acc > best_accuracy:
            best_accuracy = acc
            best_mapping = perm

    return best_accuracy, best_mapping
```
