```python
def norm(X):
    return (X-X.mean())/X.std()
```

```python
X_train=norm(X_train)
Y_train=norm(Y_train)
```

```python
X_train = X_train.to_numpy()
X_test=X_test.to_numpy()
Y_train = Y_train.to_numpy()
Y_test=Y_test.to_numpy()
```

```python
def z_calc(X,W,B):
    return np.matmul(X,W)+B

def g_calc(Z):
    return 1/(1+np.exp(-Z))

def compute_cost(Y_train,G_train):
    epsilon=1e-15
    G_train=np.clip(G_train, epsilon, 1 - epsilon)
    cost=-(Y_train*np.log(G_train)+(1-Y_train)*np.log(1-G_train))/G_train.shape[0]
    return cost.sum()
```

```python
W_train=np.ones((12,1))
B_train=np.ones(1)
```

```python
Z_train=z_calc(X_train,W_train,B_train)
G_train=g_calc(Z_train)
P_train=np.round(G_train)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(Z_train.shape)
```

```python
W_history=[]
B_history=[]
P_history=[]
J_history=[]
W_history.append(W_train)
B_history.append(B_train)
P_history.append(P_train)
J_history.append(compute_cost(Y_train,G_train))
```

```python
def gradescent(epoch, alpha, W_history, J_history, B_history, P_history, X_train, Y_train, Z_train, G_train):
    for i in range(epoch):
        W_history.append(W_history[-1]-alpha/X_train.shape[0]*X_train.T@(G_train-Y_train))
        B_history.append(B_history[-1]-alpha/X_train.shape[0]*np.sum((G_train-Y_train)))
        Z_train = z_calc(X_train, W_history[-1], B_history[-1])
        G_train = g_calc(Z_train)
        P_history.append(np.round(G_train))
        J_history.append(compute_cost(Y_train,G_train))
```

```python
epoch=1000
alpha=.1
gradescent(epoch, alpha, W_history, J_history, B_history, P_history, X_train, Y_train, Z_train, G_train)
```
